```python
df["emp_length"]=df["emp_length"].str.replace("10+ years", "10", regex=False)
df["emp_length"]=df["emp_length"].str.replace("< 1 year", "1", regex=False)
df["emp_length"]=df["emp_length"].str.replace(" years", "", regex=False)
df["emp_length"]=df["emp_length"].str.replace(" year", "", regex=False)
```

```python
df["emp_length"]=pd.to_numeric(df["emp_length"])
```

- Create our default flag and categorize for risk analysis.

```python
df["loan_status"].unique()
```

```python
def classify_risk(loan_status):
    if loan_status in ["Fully Paid", "Current"]:
        return "Performing loans"
    elif loan_status in ["In Grace Period", "Late (16-30 days)"]:
        return "Watchlist loans"
    elif loan_status in ["Late (31-120 days)"]:
        return "High risk loans"
    elif loan_status in ["Charged Off", "Default"]:
        return "Defaulted"
    else:
        return "Unknown" 
```

```python
df["risk_stage"]=df["loan_status"].apply(classify_risk)
```
